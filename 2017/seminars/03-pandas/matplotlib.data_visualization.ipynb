## Много гарфиков на одной плоскости

#### Сгенирируем данные

```python
for i in range(10):
    print (i**2)
```

```python
linear = range(10)
nonlinear = [x**2 for x in range(10)]
```

```python
nonlinear
```

#### Построим график

```python
plt.grid()
plt.plot(linear, 'b-')
plt.plot(nonlinear, 'g^')
```

#### Подписи к осям

```python
?plt.scatter(6, 6)
```

```python
plt.grid()
plt.plot(linear, 'b-')
plt.plot(nonlinear, 'g-')
plt.scatter(6, 6)
plt.legend(['linear', 'nonlinear'])
```

```python
plt.grid()
plt.plot(linear, 'b-')
plt.plot(nonlinear, 'g-')
plt.legend(['linear', 'nonlinear'], loc=5)
```

## Много графиков на одном изображении

```python
plt.figure(figsize=(6,4))

plt.subplot(211)
plt.grid()
plt.plot(linear, 'b-')

plt.subplot(212)
plt.grid()
plt.plot(nonlinear, 'g-')
```

## Как сохранить график

```python
plt.grid()
plt.plot(linear, 'b-')
plt.plot(nonlinear, 'r*')
plt.legend(['linear', 'nonlinear'], loc=2)
plt.savefig('./saved_plot.png')
```

## Задача "Ирисы Фишера"

```python
from sklearn import datasets
from pandas import DataFrame
```

```python
iris = datasets.load_iris()
```

```python
print (iris.DESCR)
```

```python
iris.data
```

```python
iris.target_names
```

```python
iris_frame = DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
iris_frame['target_name'] = iris_frame.target.apply(lambda x : iris.target_names[x])
```

```python
iris_frame.head()
```

```python
!pwd
```

```python
iris_frame.to_csv('iris_frame.csv', header=True, index=False, sep=';')
```

```python
iris_frame[(iris_frame.target > 1.5)]
```

```python
iris_frame.target = iris_frame.target.apply(int)
```

```python
iris_frame.describe()
```

```python
iris_frame[iris_frame.target == 0].hist('sepal length (cm)')
```

```python
iris_frame[iris_frame.target == 1].hist('sepal length (cm)')
```

```python
plt.figure(figsize = (20, 24))

plot_number = 0
for feature_name in iris['feature_names']:
    for target_name in iris['target_names']:
        plot_number += 1
        plt.subplot(4, 3, plot_number)
        plt.hist(iris_frame[iris_frame.target_name == target_name][feature_name])
        plt.title(target_name)
        plt.xlabel('cm')
        plt.ylabel(feature_name[:-4])
```

```python
import seaborn as sns
```

```python
sns.pairplot(iris_frame, hue='target')
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=3, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
train_predict = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(train_predict, y_train)
```

```python
predict = model.predict(X_test)
```

```python
predict
```

```python
y_test
```

```python
accuracy_score(y_test, predict)
```
